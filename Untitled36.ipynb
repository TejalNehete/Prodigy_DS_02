```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
```

```python
# Check the first few rows
print(data.head())

# Summary of the dataset
print(data.info())

# Summary statistics
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())

# Fill missing values if any (example)
# data.fillna(method='ffill', inplace=True)
```

```python
# Convert 'species' to a categorical type
data['species'] = data['species'].astype('category')
```

```python
# Distribution of Sepal Length
plt.figure(figsize=(10, 6))
sns.histplot(data['sepal length (cm)'], bins=30, kde=True)
plt.title('Sepal Length Distribution')
plt.show()

# Distribution of Species
plt.figure(figsize=(10, 6))
sns.countplot(x='species', data=data)
plt.title('Species Distribution')
plt.show()
```

```python
# Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=data)
plt.title('Sepal Length vs Sepal Width')
plt.show()

# Petal Length vs Petal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', data=data)
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
# Pair plot
sns.pairplot(data, hue='species')
plt.show()
```

```python
# Correlation matrix
corr = data.iloc[:, :-1].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
